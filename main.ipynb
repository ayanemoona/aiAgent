```python

from os import link

import requests

PROMPT = """
I have the following functions in my system:

`get_popular_movies`
`get_movie_details`
`get_movie_credits`

All functions except `get_popular_movies` require an integer movie ID as an argument.

Examples:
get_popular_movies()
get_movie_details(1234)
get_movie_credits(1234)

Please answer with only the name of the function you would like me to run, including any required arguments.

Do not provide any explanation or additional text.

Question:
movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘
"""

def get_popular_movies():
    link = os.getenv("MOVIE_API_URL") + "/movies"
    response = requests.get(link)

    return response.json()

def get_movie_credits(movie_id):
    link = f"{os.getenv('MOVIE_API_URL')}/movies/{movie_id}/credits"
    response = requests.get(link)
    print(link)

    print(response.status_code)
    print(response.text[:200])

    return response.json()

def get_movie_details(movie_id):
    link = f"{os.getenv('MOVIE_API_URL')}/movies/{movie_id}"
    response = requests.get(link)
    print(link)

    print(response.status_code)
    print(response.text[:200])
    return response.json()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": PROMPT
        }
    ]
)
result = response.choices[0].message.content
function_name = result.split("(")[0]
argument = int(result.split("(")[1].replace(")", ""))

if function_name == "get_popular_movies":
    answer = get_popular_movies()
elif function_name == "get_movie_details":
    answer = get_movie_details(argument)
elif function_name == "get_movie_credits":
    answer = get_movie_credits(argument)
answer
```
