```python
import openai

client = openai.OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": "Hello my name is moona"
        }
    ]
)
print(response.choices[0].message.content)
```

```python
import openai

client = openai.OpenAI()

messages = []
```

```python
def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    ) 
    massage = response.choices[0].message.content
    messages.append({"role": "assistant", "content": massage})
    print(f"AI: {massage}")
```

```python
while True:
    message = input("Send a message to the LLM ...")
    if message == "quit" or message == "q":
        break
    else:
        messages.append({
            "role": "user",
            "content": message
        })
        print(f"User: {message}")
        call_ai()
    
    
```
